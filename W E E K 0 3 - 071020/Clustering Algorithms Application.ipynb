```
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import Birch
```

```
X, clusters = make_blobs(n_samples=450, centers=6, cluster_std=0.70, random_state=0)
plt.scatter(data['age'], data['duration'], alpha=0.5, edgecolors='b')
```

```
age_duration_data = data[['age','duration', 'pdays']]
age_duration_data
```

```
brc = Birch(branching_factor=50, n_clusters=None, threshold=1.5)
brc.fit(age_duration_data)
```

```
labels = brc.predict(age_duration_data)
labels
```

```
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(age_duration_data['age'], age_duration_data['duration'], age_duration_data['pdays'], c=labels, cmap='rainbow', alpha=0.7, edgecolors='b')
```
